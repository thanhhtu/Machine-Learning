GradientDescent

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 100 # Số lượng mẫu

# Tạo dữ liệu X ngẫu nhiên (mỗi giá trị nằm trong khoảng từ 0 đến 10)
X = np.random.rand(m, 1) * 10  # X là một cột, có kích thước (100, 1), đã có dạng 0,1 rồi

# Giả định mối quan hệ tuyến tính y = 4 + 3*X + noise
noise = np.random.randn(m, 1) * 2  # Thêm nhiễu Gaussian vào dữ liệu
y = 4 + 3 * X + noise  # y có kích thước (100, 1)

plt.plot(X, y, 'ro')

plt.show()
```

```python
def cost(w,b):
    l=0
    for i in range(0, len(X)):
        l+=(w*X[i]+b-y[i])**2
    return l/(2*len(X))
```

```python
def gradient(w,b):
    dw=0
    db=0
    for i in range(0, len(X)):
        dw+=(w*X[i]+b-y[i])*X[i]
        db+=(w*X[i]+b-y[i])
    return dw/len(X), db/len(X)
```

```python
def train():
    w=0
    b=0
    learning_rate=0.0025
    for i in range(12000):
        dw, db=gradient(w,b)
        w=w-learning_rate*dw
        b=b-learning_rate*db
        if i % 100 == 0:
            print(f"Iteration {i}: w = {w}, b = {b}")
            
    return w, b
```

```python
w_opt, b_opt = train()

print("Optimal w:", w_opt)
print("Optimal b:", b_opt)
```

```python
x_test = 10
y_pred = w_opt * x_test + b_opt
print(y_pred)
```

```python
plt.plot(X, y, 'ro')
plt.plot([0, 10], [b_opt, w_opt*10+b_opt], color="blue", label="Đường hồi quy tuyến tính")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

Implement Linear Regression theo OOP-style

```python
class LinearRegression:
    def __init__(self,w=0,b=0):
        self.w = w
        self.b = b
    
    def cost(self, X, y):
        l=0
        for i in range(0, len(X)):
            l+=(self.w*X[i]+self.b-y[i])**2
        return l/(2*len(X))
    
    def gradient(self, X, y):
        dw=0
        db=0
        for i in range(0, len(X)):
            dw+=(self.w*X[i]+self.b-y[i])*X[i]
            db+=(self.w*X[i]+self.b-y[i])
        return dw/len(X), db/len(X)
    
    def train(self,X,y, learning_rate=0.002):
        for i in range(12000):
            dw, db=self.gradient(X, y)
            self.w=self.w-learning_rate*dw
            self.b=self.b-learning_rate*db
            if i % 100 == 0:
                print(f"Iteration {i}: w = {self.w}, b = {self.b}")
        return self.w, self.b
    
    def predict(self, x):
        return self.w*x+self.b
    
    def plot(self, X, y):
        plt.plot(X, y, 'ro')
        plt.plot([0, 10], [self.b, self.w*10+self.b], color="blue", label="Đường hồi quy tuyến tính")
        plt.xlabel("X")
        plt.ylabel("y")
        plt.show()

model = LinearRegression()
model.train(X,y)
model.plot(X, y)
    
```
